# Splitting Dataset as X and Y

```python
X_data=data.drop(columns="Y")
Y_data=data["Y"]
```

```python
X_data.shape,Y_data.shape
```

# Train Test Splitting

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_data,Y_data,test_size=0.3,stratify=Y_data,random_state=100)
```

```python
y_train.value_counts()
```

# Model Building

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
classifier = Sequential()
```

```python
classifier.add(Dense(units = 29, input_dim = 57))
classifier.add(Dense(units = 29 ))
classifier.add(Dense(units = 29))
classifier.add(Dense(units = 1,  activation = 'sigmoid'))
```

```python
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
classifier.summary()
```

# model training

```python
classifier.fit(X_train,y_train, verbose=1,validation_split=0.2,epochs=500)
```

# model evaluation

```python
classifier.evaluate(
  x=X_test,
  y=y_test,
  batch_size=None,
  verbose=1,use_multiprocessing=True
)
```

```python
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)
```

# metrics

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
